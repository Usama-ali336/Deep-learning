```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.models import Sequential
import numpy as np
import pandas as pd
```

```python
from tensorflow.keras.datasets import fashion_mnist
mnist_data = fashion_mnist.load_data()
```

```python
(x_train,y_train),(x_test,y_test) = mnist_data
```

```python
x_train.shape
```

```python
x_train = x_train/255.0
x_test = x_test/255.0
```

```python
import matplotlib.pyplot as plt
```

```python
labels = [
    'T-shirt/top',
    'Trouser',
    'Pullover',
    'Dress',
    'Coat',
    'Sandal',
    'Shirt',
    'Sneaker',
    'Bag',
    'Ankle boot'
]
print(y_train[0])
```

```python
plt.imshow(x_train[1])
plt.show()
print(f"label: {labels[y_train[1]]}")
```

```python
figure,ax = plt.subplots(1,9,figsize=(10,5))

for i in range(9):
  ax[i].set_axis_off()
  ax[i].imshow(x_train[i])
  
```

```python
def get_test_accuracy(model,x_test,y_test):
  test_loss,test_accuracy =  model.evaluate(x_test,y_test)
  print(test_accuracy)
```

```python
def get_model():
  model = Sequential([
          Conv2D(16,(3,3),input_shape=(28,28,1),activation='relu'),
          MaxPooling2D(2,2),
          Dense(128,activation='relu'),
          Flatten(),
          Dense(10,activation='softmax')
  ])
  model.compile(optimizer='Adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])

  return model
```

```python
model = get_model()
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

early_stopping = EarlyStopping(monitor='val_accuracy', mode ='max',patience=3)
```

```python
checkpoint_path = 'checkpoint_file'
checkpoint = ModelCheckpoint(filepath=checkpoint_path,
                             monitor = 'val_accuracy',
                             mode='max',
                             save_weights_only=True,
                             save_best_only=True
                             )
```

```python
history = model.fit(x_train,y_train,validation_split =0.15, epochs=20, batch_size=64, callbacks=[early_stopping,checkpoint])
```

```python
df = pd.DataFrame(history.history)
df.head()
```

```python
loss = df.plot(y='loss', title='loss vs epochs')
val_loss =  df.plot(y='val_loss')
val_loss.set(xlabel='epochs',ylabel ='loss')
```

```python
acc_plot = df.plot(y='accuracy',title='accuracy vs epochs')
val_acc_plot = df.plot(y='val_accuracy')
```
